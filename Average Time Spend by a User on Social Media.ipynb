#  **Analysing the Data through Data Visualisation**

- ## **Time spent on Social Media by different age groups**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_Average time spent by user on Social Media_data.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(12)
customerDf['time_spent'].groupby(customerDf['age']).sum().plot(kind='bar')

plt.xlabel('age')
plt.ylabel('Time Spent (in hours)')
plt.title('usage of Social Media by Different age group & Time')

plt.show()
```

- ##  **Comparing the usage of Social Media platform**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath = 'after_drop_Average time spent by user on Social Media_data.csv'
customerDf=pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure(figsize=(6,5))
mycolor=np.array(['red','lightgreen','pink'])
customerDf['platform'].value_counts().plot(kind='pie',autopct='%1.0f%%',colors=mycolor)
plt.title('Different Social Media Platforms')
plt.legend(title='Platforms')
plt.show()
```

- ##  **Analysing the usage by gender**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_Average time spent by user on Social Media_data.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(5)
mycolor=np.array(['blue','yellow','red'])
customerDf['time_spent'].groupby(customerDf['gender']).sum().plot(kind='bar', color=mycolor)
plt.xlabel('gender')
plt.ylabel('Time Spent (in hours)')
plt.title('usage of Social Media by gender')
plt.show()
```

- ##  **Time spent on various platforms**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath2 = 'after_drop_Average time spent by user on Social Media_data.csv'
customerDf=pd.read_csv(filepath2)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
plt.figure().set_figwidth(5)
mycolor=np.array(['cyan','yellow','pink'])
customerDf['time_spent'].groupby(customerDf['platform']).sum().sort_values(ascending=False).plot(kind='bar',color=mycolor)
plt.xlabel('platform')
plt.ylabel('Time Spent (in hours)')
plt.title('Usage of Different Social Media Platforms')
plt.show()
```

- ##  **Demographical distribution of Social Media Usage**

- ##  **Usage by different Location**
